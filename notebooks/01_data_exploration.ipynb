Load data, basic checks, class balance, and simple visuals.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path('../data/raw/creditcard.csv')
df = pd.read_csv(data_path)
print(df.head())
print('\nShape:', df.shape)
print('\nClass balance:\n', df['Class'].value_counts(normalize=True))

# Amount distribution (log)
import math
df['log_amount'] = (df['Amount'] + 1).apply(lambda x: math.log(x))
df['log_amount'].hist(bins=50)
plt.title('Log Amount Distribution')
plt.xlabel('log(Amount+1)')
plt.ylabel('Frequency')
plt.show()
```
